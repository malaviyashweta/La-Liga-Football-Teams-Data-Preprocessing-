```python
import pandas as pd
import numpy as np
```

```python
#reading the data set
df_laliga = pd.read_csv('Laliga.csv', skiprows = 1)   #skiprows is used as the first row is blank in the data
df_laliga.head()
```

```python
df_laliga.replace(to_replace="-",value=0, inplace=True) #replacing - with 0
df_laliga
```

```python
df_laliga.dtypes
```

```python
df_laliga["Debut"].unique
```

```python
#we can see that apart from 1929, all other years are separated by "-". So, to deal with it, we will split the Debut column in 2 columns
```

```python
new = df_laliga["Debut"].str.split("-", expand = True) 
```

```python
df_laliga["Debut_A"]= new[0]
df_laliga["Debut_B"]= new[1]
```

```python
df_laliga.dtypes
```

```python
df_laliga["Debut_A"] = df_laliga["Debut_A"].astype(int) #in order to perform arithmetical operations
```

```python
df_laliga.head()
```

```python
#Print all the teams which have started playing between 1930-1980.
df2 = df_laliga[(df_laliga['Debut_A'] >= 1930) & (df_laliga['Debut_A'] <= 1980)]  
print(df2.shape)
print(df2[['Team','Debut']])
```

```python
#Print the list of teams which came Top 5 in terms of point

df_laliga["Points"] = df_laliga["Points"].astype(int)
df_sorted = df_laliga.sort_values(by='Points', ascending=False)
df_sorted.loc[0:4,["Team","Points"]]
```

```python
# Defining the function Goal_diff_count which should return all the teams with their Goal Differences.

df_laliga["GoalsFor"] = df_laliga["GoalsFor"].astype(int)
df_laliga["GoalsAgainst"] = df_laliga["GoalsAgainst"].astype(int)

df_laliga.dtypes
def Goal_diff_count(col1, col2):
    return col1-col2

df_laliga['Goal_differences'] = Goal_diff_count(df_laliga['GoalsFor'], df_laliga['GoalsAgainst'])

df_Goal_diff = df_laliga[['Team' , 'Goal_differences']]
df_Goal_diff
```

```python
# using function Goal_diff_count to find the team which has maximum goal difference
Max_Goals = df_laliga.loc[df_laliga['Goal_differences'].idxmax()]
print("Team with max goal difference: ",Max_Goals['Team'])


# using function Goal_diff_count to find the team which has minimum goal difference
Min_Goals = df_laliga.loc[df_laliga['Goal_differences'].idxmin()]
print("Team with min goal difference: ",Min_Goals['Team'])
```

```python
df_laliga["GamesWon"] = df_laliga["GamesWon"].astype(int)
df_laliga["GamesPlayed"] = df_laliga["GamesPlayed"].astype(int)


#Creating a new column with name “Winning Percent” and appending it to the data set
df_laliga['Winning Percent'] = ((df_laliga['GamesWon']/df_laliga['GamesPlayed'])*100)
df_laliga.head()
```

```python
# replacing numerical error with 0
values = {'Winning Percent': 0}
df_laliga.fillna(value=values)
```

```python
# Printing the top 5 teams which has the highest Winning percentage
df_sorted_win_pt = df_laliga.sort_values(by='Winning Percent', ascending=False)
df_sorted_win_pt.loc[:4,"Team"]
```

```python
#Grouping teams based on their “Best position” and printing the sum of their points for all positions
Best_position_Teams=df_laliga.groupby('BestPosition', as_index= False).agg({"Points": "sum"})
Best_position_Teams
```
